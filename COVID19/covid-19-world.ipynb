# Import libraries

```python
import numpy as np
import pandas as pd
from urllib.request import urlretrieve as retrieve
```

# Download csv file and return a pandas DataFrame

```python
URL = "https://covid.ourworldindata.org/data/owid-covid-data.csv"
```

```python
def retrieve_csv_data (url = URL):
    retrieve (url, "covid_data.csv")
    return pd.read_csv("covid_data.csv")
```

```python
covid_data = retrieve_csv_data()
```
